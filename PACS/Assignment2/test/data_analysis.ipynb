# Data Analysis of the results

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_clock = pd.read_csv('./clock/clock_stats.csv')
df_gettimeofday = pd.read_csv('./gettimeofday/gettimeofday_stats.csv')
df_clock.head()
```

```python
df_gettimeofday.head()
```

```python
plt.xlabel('Matrix Size')
for label in df_clock['Executable'].unique():
    plt.plot(df_clock[df_clock['Executable'] == label]['Matrix Size'],
                df_clock[df_clock['Executable'] == label]['mean_clock_time'],
                marker='o', label=label)
    plt.plot(df_clock[df_clock['Executable'] == label]['Matrix Size'],
             df_clock[df_clock['Executable'] == label]['mean_user_time'],
             marker='o', linestyle='--', label=label)
plt.legend()
plt.ylabel('Mean Clock Time')
plt.title('Mean Clock Time')
plt.grid(True)
plt.show()
```

```python
plt.xlabel('Matrix Size')
for label in df_gettimeofday['Executable'].unique():
    plt.plot(df_gettimeofday[df_gettimeofday['Executable'] == label]['Matrix Size'],
                df_gettimeofday[df_gettimeofday['Executable'] == label]['mean_initialization_time'],
                marker='o', label=label)
plt.legend()
plt.ylabel('Initialization Time')
plt.title('Initialization Time')
plt.grid(True)
plt.show()
```

```python
plt.xlabel('Matrix Size')
for label in df_gettimeofday['Executable'].unique():
    plt.plot(df_gettimeofday[df_gettimeofday['Executable'] == label]['Matrix Size'],
             df_gettimeofday[df_gettimeofday['Executable'] == label]['mean_multiplication_time'],
             marker='o', label=label)
    plt.plot(df_gettimeofday[df_gettimeofday['Executable'] == label]['Matrix Size'],
             df_gettimeofday[df_gettimeofday['Executable'] == label]['mean_real_time'],
             marker='o', linestyle='--', label=label)
plt.legend()
plt.title('Mean Multiplication Time')
plt.ylabel('Multiplication Time')
plt.grid(True)
plt.show()
```
